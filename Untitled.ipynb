```python
import pandas as pd
import numpy as np
import re
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings; warnings.simplefilter('ignore')

df = pd.read_csv('Anapa_plane.csv')
```

```python
# Функция вывода месяца из строки
def departure_month(x):
    return datetime.strptime(x, '%Y-%m-%dT%H:%M:%SZ').month

# функция расходов на топливо в зависимости от модели самолета и месяца вылета
def anapa_loss(model, actual_time, dep_month):
    # Часовой расход топлива
    boeing = 2400
    sukhoi = 1700

    # Стоимость 1 т топлива по месяцам
    fuel_dec = 38867/1000
    fuel_jan = 41435/1000
    fuel_feb = 39553/1000
    
    if model == 'Boeing 737-300':
        if dep_month == 1:
            return round((actual_time/60 * boeing * fuel_jan), 0)
        elif dep_month == 2:
            return round((actual_time/60 * boeing * fuel_feb), 0)
        elif dep_month == 12:
            return round((actual_time/60 * boeing * fuel_dec), 0)
    if model == 'Sukhoi Superjet-100':
        if dep_month == 1:
            return round((actual_time/60 * sukhoi * fuel_jan), 0)
        elif dep_month == 2:
            return round((actual_time/60 * sukhoi * fuel_feb), 0)
        elif dep_month == 12:
            return round((actual_time/60 * sukhoi * fuel_dec), 0)
```

## Подготовка данных и создание новых признаков

```python
#Подсчет количество рейсов по различным направлениям
df.arrival_airport.value_counts()
```

```python
#Поиск пропусков
df.isna().sum()
```

```python
df.loc[df['arrival_airport'] == 'NOZ'].value_counts()
```

Так как в дата сете отсутствует информация о количестве проданных билетов по направлению NOZ, то невозможно достоверно посчитать доходность этого направления. Строки с этим значением будут удалены

```python
df = df.dropna(axis=0)

df.info()
```

```python
#Создание нового столбца с месяцем вылета
df['dep_month'] = df['departure'].apply(lambda x: departure_month(x))

#Подсчет расходов на топливо в зависимости от модели самолета и месяца вылета
df['anapa_loss'] = df.apply(lambda x: anapa_loss(
    x['model'], x['actual_time'], x['dep_month']), axis=1)

df.head()
```

```python
#Расчет прибыльности каждого рейса
df['profit'] = df.total_amount - df.anapa_loss

df.sample(5)
```

```python
#Расчет заполняемости салона
df['fillability'] = round(df.tickets_sold / df.seat_amount *100, 1)

df_new=df
df_new.sample(2)
```

```python
df_new.groupby(['arrival_airport'])['profit'].min()
```

Выше приведены самые малоприбыльные рейсы по каждому направлению. Как видно уроверь прибыльности абсолютно разный, поэтому сравнивать их между собой не имеет смысла и далее будут рассматриваться отдельно

```python
df_ego = df_new[df_new.arrival_airport == 'EGO']
df_ego.sort_values(['profit'], ascending = True)[:5]
```

По направлению Белгород было 2 рейса 136642 и 136807 с наиболее минимальной прибыльностью 473 тыс и 474 тыс соответственно. Заполняемость салона составила 66% и 70,1%

```python
df_svo = df_new[df_new.arrival_airport == 'SVO']
df_svo.sort_values(['profit'], ascending = True)[:5]
```

По направлению Москва прибыльность только 1 рейса 136352 значительно отличается от прибыльность всех последующих - 1217 тыс при заполняемости салона 70,8%
